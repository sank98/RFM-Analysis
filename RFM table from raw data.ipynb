```python
#client name
com_name = 'flyin2'
```

```python
#importing libraries

import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Read the input file

df = pd.read_csv('RFM_Transactions_'+com_name+'.csv', parse_dates = ['f0_'])
```

```python
#Converting cuid to string if required

df.loc[~df['cuid'].isnull(), 'cuid'] = df.loc[~df['cuid'].isnull(), 'cuid'].astype(str)
```

```python
df.head(10)
```

```python
#checking the number of missing values for each column

null_columns=df.columns[df.isnull().any()]
df[null_columns].isnull().sum()
```

```python
#Defining id column as: id = if cuid not null then cuid else luid

df['id'] = df[df['cuid'].isnull()]['luid']
df.loc[df['id'].isnull(), 'id'] = df.loc[df['id'].isnull(), 'cuid']
```

```python
df.info()
```

```python
#Defining hist column: Number of days between current date and transaction date

from datetime import date

df['hist']= date.today() - df['f0_'].dt.date
df['hist'].astype('timedelta64[D]')
df['hist']=df['hist'] / np.timedelta64(1, 'D')
df.head()
```

```python
#drop the cuid and luid columns

df.drop(['cuid', 'luid'], axis=1, inplace=True)
```

```python
#check for minimum and maximum hist values and report is it is unusual

print(df['hist'].min(), df['hist'].max())
```

```python
df.head(40)
```

```python
#Average Revenue

df['Revenue'].mean()
```

```python
#Fill the missing values of Revenue on the basis of groupby using id, if group not found then fill with the average Revenue

df['Revenue'] = df.groupby("id").transform(lambda x: x.fillna(x.mean()))
df.loc[df['Revenue'].isnull(), 'Revenue'] = df['Revenue'].mean()
```

```python
df.info()
```

```python
#Creating RFM Table with the given definitions:
#Groupby id
#Recency = minimum hist value (number of days before most recent transaction)
#Frequency = Length of id group (Number of transaction made by the given id)
#Monetary = Sum of the revenues for all transaction of the given id

import statistics

rfmTable = df.groupby('id').agg({'hist': lambda x:x.min(),                    # Recency
                                        'id': lambda x: len(x),               # Frequency
                                        'Revenue': lambda x: x.sum(),         # Monetary Value
                                'f0_': lambda x: x.max(),       
                                'weekday': lambda x: sum(x)})          


rfmTable.rename(columns={'hist': 'recency', 
                         'id': 'frequency', 
                         'Revenue': 'monetary_value',
                        'f0_': 'date'}, inplace=True)
```

```python
#round the recency column

rfmTable['recency'] = round(rfmTable['recency'])
```

```python
#Save the RFM table for further use

rfmTable.to_csv("rfmTable_"+com_name+".csv", sep='\t', encoding='utf-8')
```
